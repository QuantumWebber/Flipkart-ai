```python
import nltk
import re
```

```python
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
df = pd.read_csv(r"C:\Users\Jatin\OneDrive\Desktop\DL projects\ecom-ai\data\flipkart_product.csv", encoding="latin1")
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def preprocess(text):
    # Lowercase
    text = text.lower()
    # Special characters remove
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Extra spaces remove
    text = re.sub(r'\s+', ' ', text).strip()
    # Stopwords remove
    tokens = text.split()
    tokens = [w for w in tokens if w not in stop_words]
    return ' '.join(tokens)
```

```python
def preprocess(text):
    if not isinstance(text, str):
        return ''
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = text.split()
    tokens = [w for w in tokens if w not in stop_words]
    return ' '.join(tokens)

df['clean_summary'] = df['Summary'].apply(preprocess)
```

```python
df = df[df['clean_summary'].str.len() > 0]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Rate'] = pd.to_numeric(df['Rate'], errors='coerce')
df = df.dropna(subset=['Rate'])
df['sentiment'] = df['Rate'].apply(lambda x: 'positive' if x >= 4 else ('negative' if x <= 2 else 'neutral'))
```

```python
df.shape
```

```python
df["sentiment"].value_counts()
```

```python
df.isnull().sum()
```

```python
from sklearn.utils import resample
```

```python
positive = df[df['sentiment'] == 'positive']
negative = df[df['sentiment'] == 'negative']
neutral = df[df['sentiment'] == 'neutral']
```

```python
positive.shape
```

```python
neutral.shape
```

```python
#class imbalance handle
```

```python
min_count=min(len(positive),len(negative),len(neutral)
              
              )
```

```python
min_count
```

```python
positive_balanced = resample(positive, n_samples=min_count, random_state=42)
negative_balanced = resample(negative, n_samples=min_count, random_state=42)
neutral_balanced = resample(neutral, n_samples=min_count, random_state=42)
```

```python
df_balanced = pd.concat([positive_balanced, negative_balanced, neutral_balanced])
```

```python
df_balanced.shape
```

```python
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_balanced
```
